```python
from google.colab import files
import pandas as pd
import io
```

```python
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]))
```

```python
df
```

```python
# Assuming df is your DataFrame
columns_to_keep = ['student name', 'student sis', 'section', 'final score']

# Keep only these columns
df = df[columns_to_keep]

# Display the cleaned DataFrame
df
```

```python
# Filter the DataFrame to keep only rows where 'section' matches
filtered_df = df[df['section'] == 'MKTG563_55FA25']

# Reset index (optional)
filtered_df = filtered_df.reset_index(drop=True)

# Display the filtered DataFrame
filtered_df.head()
```

```python
# Keep only the part before '_' in the 'section' column
filtered_df['section'] = filtered_df['section'].str.split('_').str[0]

# Display the updated filtered DataFrame
filtered_df.head()
```

```python
# Upload a new Excel file
uploaded = files.upload()

# Get the file name (automatically picks the uploaded file)
file_name = list(uploaded.keys())[0]
print(f"Uploaded file: {file_name}")

# Load the Excel file into a DataFrame
new_df = pd.read_excel(file_name)

# Display the new DataFrame
new_df.head()
```

```python
# Ensure 'student sis' is treated as string
new_df['student sis'] = new_df['student sis'].astype(str)

# Remove leading 'M' only
new_df['student sis'] = new_df['student sis'].str.lstrip('M')

# Display the updated DataFrame
new_df.head()
```

```python
# Make sure 'student sis' and 'section' are strings in both datasets
filtered_df['student sis'] = filtered_df['student sis'].astype(str)
filtered_df['section'] = filtered_df['section'].astype(str)
new_df['student sis'] = new_df['student sis'].astype(str)
new_df['section'] = new_df['section'].astype(str)

# Merge new_df with filtered_df on 'student sis' and 'section'
merged_df = new_df.merge(
    filtered_df[['student sis', 'section', 'final score']],  # only keep necessary columns from filtered_df
    on=['student sis', 'section'],
    how='left'  # keeps all students from new_df; adds final score if match found
)

# Display the merged DataFrame
merged_df
```

```python
# Convert 'WEEK 5' column to numeric (coerce errors to NaN)
merged_df['WEEK 5'] = pd.to_numeric(merged_df['WEEK 5'], errors='coerce')

# Count students with score less than 80
count_below_80 = merged_df[merged_df['WEEK 5'] < 80].shape[0]

print(f"Number of students with WEEK 5 score less than 80: {count_below_80}")
```

```python
# Save the final merged DataFrame to Excel
merged_df.to_excel('final_merged.xlsx', index=False)

# Download the Excel file
files.download('final_merged.xlsx')
```
